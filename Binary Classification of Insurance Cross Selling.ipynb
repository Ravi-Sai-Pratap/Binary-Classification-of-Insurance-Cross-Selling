```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, AdaBoostClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler

def analyze_discrete_target(train_csv, test_csv, target_col='Response', test_size=0.2, random_state=42, submission_file='submission.csv'):
    """
    Perform analysis on a discrete target variable in provided CSV files and generate predictions for the test set.

    Parameters:
    - train_csv: Path to the training CSV file
    - test_csv: Path to the testing CSV file
    - target_col: Name of the target column (string)
    - test_size: Proportion of the dataset to include in the test split (float)
    - random_state: Random state for reproducibility (int)
    - submission_file: Name of the submission file (string)

    Returns:
    - None (prints results and visualizations, and saves the submission file)
    """
    # Read training and test data from CSV files
    df_train = pd.read_csv(train_csv)
    df_test = pd.read_csv(test_csv)

    # Strip trailing spaces from column names
    df_train.columns = df_train.columns.str.strip()
    df_test.columns = df_test.columns.str.strip()

    # Separate target variable for the training data
    y = df_train[target_col]

    # Drop target column from the training dataframe
    X_train = df_train.drop(columns=[target_col])

    # Separate numerical and categorical columns
    numerical_features = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()
    categorical_features = X_train.select_dtypes(include=['object']).columns.tolist()

    # One-hot encode categorical variables
    X_train_encoded = pd.get_dummies(X_train, columns=categorical_features, drop_first=True)
    X_test_encoded = pd.get_dummies(df_test, columns=categorical_features, drop_first=True)

    # Align the train and test dataframes by columns
    X_train_encoded, X_test_encoded = X_train_encoded.align(X_test_encoded, join='left', axis=1, fill_value=0)

    # Check if numerical features exist
    if numerical_features:
        # Standardization (if numerical features are normally distributed)
        scaler_standard = StandardScaler()
        X_train_encoded[numerical_features] = scaler_standard.fit_transform(X_train_encoded[numerical_features])
        X_test_encoded[numerical_features] = scaler_standard.transform(X_test_encoded[numerical_features])

    # Split scaled data into training and validation sets
    X_train_split, X_val, y_train, y_val = train_test_split(X_train_encoded, y, test_size=test_size, random_state=random_state)

    # Initialize models
    models = {
        'Logistic Regression': LogisticRegression(random_state=random_state),
        'Decision Tree': DecisionTreeClassifier(random_state=random_state),
        'Random Forest': RandomForestClassifier(random_state=random_state),
        'Bagging (Decision Tree)': BaggingClassifier(estimator=DecisionTreeClassifier(), random_state=random_state),
        'AdaBoost (Decision Tree)': AdaBoostClassifier(estimator=DecisionTreeClassifier(), random_state=random_state)
    }

    # Lists to store model names and their corresponding accuracy scores
    model_names = []
    accuracy_scores = []

    best_model = None
    best_accuracy = 0

    # Iterate through models
    for model_name, model in models.items():
        print(f"Training {model_name}...")
        
        # Fit the model
        model.fit(X_train_split, y_train)

        # Predictions
        y_pred = model.predict(X_val)

        # Model evaluation
        accuracy = accuracy_score(y_val, y_pred)
        accuracy_scores.append(accuracy)
        model_names.append(model_name)

        print(f"\nResults for {model_name}:")
        print("Confusion Matrix:\n", confusion_matrix(y_val, y_pred))
        print("\nClassification Report:\n", classification_report(y_val, y_pred))
        print("Accuracy Score:", accuracy)

        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_model = model

        # Plot confusion matrix
        plt.figure(figsize=(5, 4))
        sns.heatmap(confusion_matrix(y_val, y_pred), annot=True, fmt='d', cmap='Blues', cbar=False)
        plt.title(f'Confusion Matrix - {model_name}')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()

        print("-------------------------------------------------------")

    # Plotting accuracy scores
    plt.figure(figsize=(10, 6))
    sns.barplot(x=model_names, y=accuracy_scores)
    plt.title('Accuracy Scores of Models')
    plt.ylim(0, 1.0)
    plt.xticks(rotation=45)
    plt.ylabel('Accuracy')
    plt.xlabel('Model')
    plt.show()

    # Predict the target variable for the test dataset using the best model
    test_predictions = best_model.predict(X_test_encoded)

    # Create the submission DataFrame
    submission = pd.DataFrame({'id': df_test['id'], target_col: test_predictions})

    # Save the submission file
    submission.to_csv(submission_file, index=False)
    print(f"Submission file saved as {submission_file}")

# Example usage:
# Assuming 'train.csv' and 'test.csv' are paths to your CSV files
analyze_discrete_target('train.csv', 'test.csv', target_col='Response')
```
